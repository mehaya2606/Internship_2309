```python
#Q1 - 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos 
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
Rank = []
Name = []
Artist = []
Upload_Date = []
Views= []
```

```python
rank=driver.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
for i in rank:
    Rank.append(i.text)
    
name=driver.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
for i in name:
    Name.append(i.text)
    
artist=driver.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
for i in artist:
    Artist.append(i.text)
    
upload_date=driver.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
for i in upload_date:
    Upload_Date.append(i.text)
    
views=driver.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
for i in views:
    Views.append(i.text)
```

```python
print(len(Rank), len(Name), len(Artist), len(Upload_Date), len(Views))
```

```python
df = pd.DataFrame({"Rank":Rank, "Video Name":Name, "Artist/Uploader": Artist, "Upload Date": Upload_Date, "Views(Billion)": Views})
df
```

```python
Dataset_Name = []
Dataset_Type = []
Task= []
Attribute_Type = []
No_Of_Instances = []
No_Of_Attribute = []
Year = []

name=driver.find_elements(By.XPATH, '//h2[@class="truncate text-primary"]')
for i in name:
    Dataset_Name.append(i.text)
    
type1=driver.find_elements(By.XPATH, '//div[@class="rounded-box bg-base-100"]/div[1]/div[2]/div/div[2]/span')
for i in type1:
    Dataset_Type.append(i.text)
    
task=driver.find_elements(By.XPATH, '//div[@class="rounded-box bg-base-100"]/div[1]/div[2]/div/div[1]/span')
for i in task:
    Task.append(i.text)
    
type2=driver.find_elements(By.XPATH, '//div[@class="rounded-box bg-base-100"]/div[2]/div/table/tbody/tr/td[2]')
for i in type2:
    Attribute_Type.append(i.text)
    
instances=driver.find_elements(By.XPATH, '//div[@class="rounded-box bg-base-100"]/div[1]/div[2]/div/div[3]/span')
for i in instances:
    No_Of_Instances.append(i.text)
    
attribute=driver.find_elements(By.XPATH, '//div[@class="rounded-box bg-base-100"]/div[1]/div[2]/div/div[4]/span')
for i in attribute:
    No_Of_Attribute.append(i.text)
    
year=driver.find_elements(By.XPATH, '//div[@class="rounded-box bg-base-100"]/div[2]/div/table/tbody/tr/td[3]')
for i in year:
    Year.append(i.text)
```

```python
#Q4 - 4. Scrape the details of trending repositories on Github.com.
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://github.com/")
time.sleep(15)
```

```python
click_open_source= driver.find_element(By.XPATH, '/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
click_open_source.click()

click_trending = driver.find_element(By.XPATH, '/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')
click_trending.click()
```

```python
product_opening_url = []
url = driver.find_elements(By.XPATH, '//h2[@class="h3 lh-condensed"]/a')
for i in url:
    product_opening_url.append(i.get_attribute("href"))
```

```python
product_opening_url
```

```python
Repository_Title = []
Repository_Description = []
Contributors_Count = []
Language_Used = []

for i in product_opening_url:
    driver.get(i)
    time.sleep(5)
    try:
        title=driver.find_element(By.XPATH, '//span[@class="author flex-self-stretch"]')
        Repository_Title.append(title.text)
    except NoSuchElementException:
        Repository_Title.append('-')
        
    try:
        description=driver.find_element(By.XPATH, '//div[@class="BorderGrid BorderGrid--spacious"]/div[1]/div/div/p')
        Repository_Description.append(description.text)
    except NoSuchElementException:
        Repository_Description.append('-')
        
    try:
        count=driver.find_element(By.XPATH, '//div[@class="BorderGrid-row"][5]/div/h2/a/span')
        Contributors_Count.append(count.text)
    except NoSuchElementException:
        Contributors_Count.append('-')
        
    try:
        language=driver.find_element(By.XPATH, '//li[@class="d-inline"][1]/a/span[1]')
        Language_Used.append(language.text)
    except NoSuchElementException:
        Language_Used.append('-')
```

```python
print(len(Repository_Title), len(Repository_Description), len(Contributors_Count), len(Language_Used))
```

```python
Trending_Repository = pd.DataFrame({})
Trending_Repository['Repository Title'] = Repository_Title
Trending_Repository['Repository Description'] = Repository_Description
Trending_Repository['Contributors Count'] = Contributors_Count
Trending_Repository['Language Used'] = Language_Used
Trending_Repository
```

```python
driver.close()
```

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.shine.com/")
```

```python
designation = driver.find_element(By.CLASS_NAME, "form-control  ")
designation.send_keys("Data Analyst")
```

```python
location= driver.find_element(By.XPATH, "/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys("Bangalore")
```

```python
search = driver.find_element(By.XPATH, "/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[2]/div/button")
search.click()
```

```python
job_title = []
job_location = []
company_name = []
experience_required = []
```

```python
title_tags=driver.find_elements(By.XPATH, '//h2[@itemprop="name"]')
for i in title_tags:
    title=i.text
    job_title.append(title)
```

```python
company_tags=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]/span')
for i in company_tags:
    company=i.text
    company_name.append(company)
```

```python
experience_tags=driver.find_elements(By.XPATH, '//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experience_tags:
    experience=i.text
    experience_required.append(experience)
```

```python
location_tags=driver.find_elements(By.XPATH, '//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in location_tags:
    location=i.text
    job_location.append(location)
```

```python
print(len(company_name), len(experience_required), len(job_location), len(job_title))
```

```python
df = pd.DataFrame({"Job_Title":job_title, "Job_Location":job_location, "Company_Name": company_name, "Required_Experience": experience_required})
df
```

```python
Final_result = df.head(10)
Final_result
```
